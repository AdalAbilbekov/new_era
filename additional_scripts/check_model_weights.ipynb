```python
import numpy as np

# Your token lists
tokens_p = [1, 22, 3, 3, 4, 6]
tokens_l = [4, 6]

# Step 1: Find positions in tokens_p where tokens_l tokens are located
positions_to_mask = [i for i, token in enumerate(tokens_p) if token in tokens_l]

# Step 2: Determine the number of rows for the mask matrix
N = len(positions_to_mask) + 1  # Number of rows in the mask matrix
seq_len = len(tokens_p)         # Length of tokens_p

# Step 3: Initialize a mask of all True values
mask = np.ones((N, seq_len), dtype=bool)

# Step 4: Set positions to False progressively
for i in range(N):
    indices_to_false = positions_to_mask[i:]
    mask[i, indices_to_false] = False

# Output the mask
print(mask)
```
